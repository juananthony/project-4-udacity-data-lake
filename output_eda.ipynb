## Check with Spark

```python
import os
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
artist_df = spark.read.parquet(os.path.join(output_path, artist_path))
song_df = spark.read.parquet(os.path.join(output_path, song_path))
```
